```python
#HANDLING MISSING DATA
cars = r'C:\Users\moses\Desktop\data\cars.csv'
cars = pd.read_csv(cars)
cars.head()
```

```python
#Rename the unnamed column
cars.rename(columns = {'Unnamed: 0': 'car_features'}, inplace = True)
cars.head()
```

```python
#check the information of data
cars.info()
```

```python
cars.isna().sum()
```

```python
#Suppose you want to fill the NAs for multiple features
#Fill the NAs for a variable(eg mpg) with its mean
ave = cars['mpg'].mean()
ave
cars['mpg'].fillna(value = ave, inplace = True)
cars.isnull().sum()
```

```python
values = {'mpg':0,'cyl':0,'disp':0,'hp':0,'drat':0,'wt':0,'qsec':0,'vs':0,'am':0,'gear':0,'carb':0}
cars.fillna(value = values, inplace = True)
cars.isnull().sum()
```

```python
#Suppose we want to fill the Nas of 'disp' variable with zero
cars['disp'].fillna(value = 0, inplace = True)
cars.isnull().sum()
```

```python
#Select the gear variable with it median
gear_median = cars['gear'].median()
gear_median
cars['gear'].fillna(value = gear_median, inplace = True)
cars.isnull().sum()
```

```python
#INTRODUCTION TO MATPLOTLIB AND SEABORN
import matplotlib.pyplot as plt
import seaborn as sns

anscombe = sns.load_dataset('anscombe')
anscombe.head()
```

```python
#LINE PLOT
plt.plot(anscombe['x'], anscombe['y'])
plt.show()
```

```python
#if i need a scatter plot-SCATTER PLOT IS USE TO VISUALIZE 2 QUANTITATIVE(NUMERIC) VARIABLE
plt.plot(anscombe['x'], anscombe['y'], 'o')
plt.show()
```

```python
#how to plot a scatter plot of x and y with labels
plt.scatter(anscombe['x'], anscombe['y'])
plt.show()
```

```python
#Suppose we want to plot x and y for the dataset I
#first, select only the dataset I
data_I = anscombe[anscombe['dataset'] == 'I']
data_I
plt.scatter(data_I['x'],data_I['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('SCATTER PLOT OF X AND Y')
plt.show()
```

```python
#BAR GRAPH-This is use to count discreet(categorical variable) values
plt.bar(anscombe['dataset'], anscombe['x'])
plt.show()
```

```python
#USING SUBPLOT
data_1 = anscombe[anscombe['dataset'] =='I']
data_2 = anscombe[anscombe['dataset'] =='II']
data_3 = anscombe[anscombe['dataset'] =='III']
data_4 = anscombe[anscombe['dataset'] =='IV']
fig = plt.figure(figsize = (10,5))
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)

#to plot on the axes
axes1.plot(data_1['x'], data_1['y'], 'o')
axes2.plot(data_2['x'], data_2['y'], 'ro')

#suppose i want to set the title
axes1.set_title('dataset I')
axes2.set_title('dataset II')

#Add a title for the entire figure
fig.suptitle('Anscombe Data')

#Using a tight layout
fig.tight_layout()
```

```python
#ALTERNATIVE METHOD
fig, ax = plt.subplots(2,2)
ax[0,0].plot(data_1['x'], data_1['y'], 'o')
ax[0,0].set_title('Dataset I')

ax[0,1].plot(data_2['x'], data_2['y'], 'ro')


ax[1,0].plot(data_3['x'], data_3['y'], 'go')


ax[1,1].plot(data_4['x'], data_4['y'], 'o')


plt.show()
```

```python
#HISTOGRAM: This is used to visualize the distribution of a single quantitative variable.
tips = sns.load_dataset('tips')
tips.head()
```

```python
#PLOT A HISTOGRAM OF TOTAL BILL
plt.hist(tips['total_bill'])
plt.xlabel('Frequency')
plt.ylabel('Total Bills')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
#ALTERNATIVE METHOD
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.hist(tips['total_bill'])
axes.set_xlabel('Frequency')
axes.set_ylabel('Total Bill')
axes.set_title('Histogram of Total Bill')
plt.show()
```
